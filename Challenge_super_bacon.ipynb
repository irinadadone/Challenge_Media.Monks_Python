```python
# concat two headers in one and delete Unnamed column names
# return an array with the final row names
def concat_header(row1, row2):
    finalArray = []
    for i in range(len(row1)):
        if 'Unnamed' in row1[i]:
            finalArray.append(row2[i])
        else:
            finalArray.append(row1[i] + row2[i])
    return finalArray
```

```python
superbacon_header = concat_header(superbacon_column_names, superbacon_column_names2)
print(superbacon_header)
```
